```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data=pd.read_csv("/home/banga/Desktop/datamining/beta/beta_data_stat.csv")
```

```python
data2=data[0:104]
```

```python
data5={}
```

```python

data5["Equity"]=np.sum(np.square(data2.Equity-data2.Equity.mean()))/104
data5["KCB"]=np.sum(np.square(data2.KCB-data2.KCB.mean()))/104
data5["UCHUMI"]=np.sum(np.square(data2.UCHUMI-data2.UCHUMI.mean()))/104
data5["BAMBURI"]=np.sum(np.square(data2.BAMBURI-data2.BAMBURI.mean()))/104
data5["BAT"]=np.sum(np.square(data2.BAT-data2.BAT.mean()))/104
data5["SASINI"]=np.sum(np.square(data2.SASINI-data2.SASINI.mean()))/104
data5["SAFAR"]=np.sum(np.square(data2.SAFAR-data2.SAFAR.mean()))/104
data5["NMG"]=np.sum(np.square(data2.NMG-data2.NMG.mean()))/104
data5["SCB"]=np.sum(np.square(data2.SCB-data2.SCB.mean()))/104
data5["EABL"]=np.sum(np.square(data2.EABL-data2.EABL.mean()))/104
```

```python
d2=pd.Series(data5)
d2.to_csv("/home/banga/Desktop/datamining/beta/mean-squarederror.csv")
```

```python
fig, ax= plt.subplots(figsize=(20,5))

ax.plot(data5.index, data5.Equity)
ax.set_xlabel('Time (Weeks)')
ax.set_ylabel('Equity ABS')
ax.set_title('Equity ABS  Vs Time');
fig.savefig("/home/banga/Desktop/datamining/beta/equityabs.png")
```
